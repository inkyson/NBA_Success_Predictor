```python
# Add the dependencies
import pandas as pd
import os
```

```python
# Files to load
stats_season_df = pd.read_csv("Resources/Stats_Ht_Season_2003-2019.csv")
player_bio_df = pd.read_csv("Resources/Player_bios.csv")
```

```python
stats_season_df
```

```python
# Drop unnecessary columns from stats_season
stats_season_df = stats_season_df.drop(columns=['Rk', 'Ht', 'Lg'])
```

```python
stats_season_df.head()
```

```python
# Remove accents from Player names
stats_season_df['Player'] = stats_season_df['Player'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
# Remove periods from all names
stats_season_df.Player=stats_season_df.Player.str.replace('.','')
```

```python
stats_season_df
```

```python
stats_season_df.to_csv('Resources/stats_season_clean.csv', index=False)
```

```python
player_bio_df
```

```python
# Drop unnecessary columns
player_bio_df = player_bio_df.drop(columns=['Unnamed: 0'])
```

```python
# Remove periods from all names
player_bio_df.player_name=player_bio_df.player_name.str.replace('.','')
```

```python
# Add each suffix to remove to a list.
suffixes = [" Jr."," Sr.", " Jr", " III", " II", " IV", " Sr"]

# Iterate through the words in the "suffixes" list and replace them with an empty space, "".
for word in suffixes:
    player_bio_df["player_name"] = player_bio_df["player_name"].str.replace(word,"")
```

```python
player_bio_df.head()
```

```python
# Drop rows from years before 2003-04
player_bio_df = player_bio_df.drop(player_bio_df[player_bio_df.season.isin(["1996-97", "1997-98", "1998-99", "1999-00", "2000-01", "2001-02", "2002-03"])].index)
```

```python
player_bio_df.to_csv('Resources/player_bio_clean.csv', index=False)
```

```python
nba_players_df = pd.merge(stats_season_df, player_bio_df,  how='left', left_on=['Player','Season'], right_on = ['player_name','season'])
```

```python
nba_players_df
```

```python
# Determine if there are any missing values
nba_players_df.isnull().sum()
```

```python
nba_players_df.to_csv('Resources/nba_players.csv', index=False)
```
